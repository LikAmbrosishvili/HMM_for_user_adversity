```python
import numpy as np 
import matplotlib.pyplot as plt

sim = 1000000

A = np.random.uniform(1,5, sim)
B = np.random.uniform(2,6, sim)


duration  = A + B
duration.shape
```

```python
plt.figure(figsize = (3, 1.5))
plt.hist(duration, density =True)
plt.axvline(9, color = 'r')
plt.show()
print((duration>9).sum()/sim)
```
